# Exploring Categorical Data - BMI Data Analysis with Visualizations

This notebook demonstrates how to analyze and visualize BMI data using Python. It includes examples of:

- Creating and displaying DataFrames
- Calculating percentages
- Creating bar charts (grouped and stacked)
- Creating a pie chart

```python
# Create a DataFrame with the BMI ranges and categories
bmi_data = {
    "BMI range": ["< 18.5", "18.5â€“24.9", "25.0â€“29.9", "30.0 and above"],
    "Category": ["Underweight", "Normal weight", "Overweight", "Obese"]
}

bmi_df = pd.DataFrame(bmi_data)
```

```python
bmi_df
```

```python
bmi_df["Sample Count"] = [25, 500, 175, 50]  # Hypothetical values
total_count = bmi_df["Sample Count"].sum()

# Calculate percentage
bmi_df["Percentage"] = (bmi_df["Sample Count"] / total_count) * 100
bmi_df["Percentage"] = bmi_df["Percentage"].round(1)
```

```python
bmi_df
```

```python
sns.set_context("talk")
```

```python
# Create a bar chart of sample counts by BMI category
with plt.style.context('fivethirtyeight'):

    plt.figure(figsize=(10, 4.5))
    plt.bar(bmi_df["Category"], bmi_df["Sample Count"])
    plt.xlabel("BMI Category")
    plt.ylabel("Count")
    plt.title("BMI Categories for Students\nClass, 2025")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.grid(axis='x')

    plt.show()
```

```python
# Create a bar chart of sample counts by BMI category
with plt.style.context('fivethirtyeight'):

    plt.figure(figsize=(10, 4.5))
    plt.barh(bmi_df["Category"], bmi_df["Sample Count"])
    plt.ylabel("BMI Category")
    plt.xlabel("Count")
    plt.title("BMI Categories for Students\nClass, 2025")
    # plt.xticks(rotation=45)
    plt.tight_layout()
    plt.grid(axis='y')

    plt.show()
```

```python
# Create a bar chart of sample counts by BMI category
with plt.style.context('fivethirtyeight'):

    plt.figure(figsize=(10, 4.5))
    plt.bar(bmi_df["Category"], bmi_df["Percentage"])
    plt.xlabel("BMI Category")
    plt.ylabel("Percentage (%)")
    plt.title("BMI Categories for Students\nClass, 2025")
    # plt.xticks(rotation=45)
    plt.tight_layout()
    plt.grid(axis='x')

    plt.show()
```

```python
# BMI data across three years
data = {
    "BMI range": [
        "Underweight",
        "Normal",
        "Overweight",
        "Obese",
        "Total"
    ],
    "2015 Count": [50, 400, 100, 10, 560],
    "2015 %": ["8.9%", "71.4%", "17.9%", "1.8%", "100.0%"],
    "2020 Count": [45, 450, 130, 40, 665],
    "2020 %": ["6.8%", "67.7%", "19.5%", "6.0%", "100.0%"],
    "2025 Count": [25, 500, 175, 50, 750],
    "2025 %": ["3.3%", "66.7%", "23.3%", "6.7%", "100.0%"]
}

bmi_df = pd.DataFrame(data)
bmi_df
```

```python
# Plot grouped bar chart
with plt.style.context('fivethirtyeight'):

    plot_df = bmi_df.iloc[:4].copy()
    plot_df["2025 %"] = plot_df["2025 %"].str.rstrip('%').astype(float)
    plot_df["2020 %"] = plot_df["2020 %"].str.rstrip('%').astype(float)
    plot_df["2015 %"] = plot_df["2015 %"].str.rstrip('%').astype(float)

    x = np.arange(len(plot_df["BMI range"]))
    width = 0.25

    # plt.figure(figsize=(10, 6))
    plt.figure(figsize=(10, 4.5))
    plt.bar(x - width, plot_df["2015 %"], width, label="2015")
    plt.bar(x, plot_df["2020 %"], width, label="2020")
    plt.bar(x + width, plot_df["2025 %"], width, label="2025")

    plt.xlabel("BMI Category")
    plt.ylabel("Percentage (%)")
    plt.title("BMI Distribution in Three Entering Classes")
    plt.xticks(x, plot_df["BMI range"]) #, rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.grid(axis='x')
    plt.show()
```

```python
with plt.style.context('fivethirtyeight'):
    years = ['2015', '2020', '2025']
    underweight = plot_df["2015 %"].values[0], plot_df["2020 %"].values[0], plot_df["2025 %"].values[0]
    normal = plot_df["2015 %"].values[1], plot_df["2020 %"].values[1], plot_df["2025 %"].values[1]
    overweight = plot_df["2015 %"].values[2], plot_df["2020 %"].values[2], plot_df["2025 %"].values[2]
    obese = plot_df["2015 %"].values[3], plot_df["2020 %"].values[3], plot_df["2025 %"].values[3]

    x = np.arange(len(years))

    plt.figure(figsize=(10, 4.5))
    plt.bar(x, underweight, label='Underweight')
    plt.bar(x, normal, bottom=underweight, label='Normal')
    plt.bar(x, overweight, bottom=np.array(underweight) + np.array(normal), label='Overweight')
    plt.bar(x, obese,
            bottom=np.array(underweight) + np.array(normal) + np.array(overweight),
            label='Obese')

    plt.xticks(x, years)
    plt.xlabel("Year")
    plt.ylabel("Percentage (%)")
    plt.title("BMI Category Distribution (Stacked) by Year")
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
    plt.tight_layout()
    plt.grid(axis='x')
    plt.show()
```

```python
labels = plot_df["BMI range"]
sizes = plot_df["2015 %"]

plt.figure(figsize=(10, 4.5))
wedges, texts, autotexts = plt.pie(
        sizes,
        # labels=labels,
        autopct='%1.0f%%',
        # labeldistance=1.3,     # Pushes labels outward
        pctdistance=1.3,
        startangle=180)
plt.title("BMI Category Distribution for Students\nClass, 2015")

plt.legend(wedges, labels, title="BMI Categories", loc="center left", bbox_to_anchor=(1, 0.5))

plt.axis('equal')
plt.show()
```

```python
# Re-import necessary libraries due to code execution environment reset
import matplotlib.pyplot as plt
import numpy as np

# Generate x values
x = np.linspace(0, 2.7 * np.pi, 300)

# Generate multiple sine wave lines with phase shifts
# lines = [np.sin(x + i) for i in range(5)]
# lines = [np.sin(x + i * 0.2) for i in range(5)]
lines = [(1 + 0.2 * i) * np.sin(x + i * 0.2) for i in range(5)]
```

```python
# Recreate the figure using the 'viridis' colormap with narrow and wide color selections
from matplotlib.cm import get_cmap

# Get the 'viridis' colormap
viridis = get_cmap('viridis')

# Select 5 close-together colors from the start of the colormap (narrow range)
narrow_colors = [viridis(i) for i in np.linspace(0.1, 0.3, 5)]

# Select 5 colors spread across the entire colormap (wide range)
wide_colors = [viridis(i) for i in np.linspace(0, 1, 5)]

# Create subplots

with plt.style.context('fivethirtyeight'):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4), sharey=True)
    # Plot narrow color range
    for i in range(5):
        ax1.plot(x, lines[i],
                linewidth=3,
                color=narrow_colors[i],
                label=f"Signal {chr(65 + i)}"   # Signal A, Signal B, ...
    )
    ax1.set_title("Narrower Color Span")
    ax1.legend(loc='lower right')

    # Plot wide color range
    for i in range(5):
        ax2.plot(x, lines[i], linewidth=3,
                color=wide_colors[i], label=f"Signal {chr(65 + i)}"   # Signal A, Signal B, ...
    )
    ax2.set_title("Wider Color Span")
    ax2.legend(loc='lower right')

    ax1.set_xlim(0, 2.7 * np.pi)
    ax2.set_xlim(0, 2.7 * np.pi)

    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# Generate synthetic 2D data
x = np.random.normal(0, 1, 1000)
y = np.random.normal(0, 1, 1000)

# Create a 2D density estimate
xy = np.vstack([x, y])
kde = gaussian_kde(xy)
xgrid = np.linspace(-3, 3, 100)
ygrid = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(xgrid, ygrid)
Z = kde(np.vstack([X.ravel(), Y.ravel()])).reshape(X.shape)

with plt.style.context('fivethirtyeight'):

    # Plot side-by-side heatmaps
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

    # First heatmap with 'jet' colormap
    im1 = ax1.imshow(Z, extent=[-3, 3, -3, 3], origin='lower', cmap='jet')
    ax1.set_title('Misleading')
    ax1.set_xlabel('Feature 1')
    ax1.set_ylabel('Feature 2')
    plt.colorbar(im1, ax=ax1)

    # Second heatmap with 'Purples' colormap
    im2 = ax2.imshow(Z, extent=[-3, 3, -3, 3], origin='lower', cmap='Purples')
    ax2.set_title('Improved')
    ax2.set_xlabel('Feature 1')
    ax2.set_ylabel('Feature 2')
    plt.colorbar(im2, ax=ax2)

    ax2.spines['bottom'].set_color('black')
    ax2.spines['left'].set_color('black')
    ax2.spines['bottom'].set_linewidth(1)
    ax2.spines['left'].set_linewidth(1)

    # hide gridlines
    ax1.grid(False)
    ax2.grid(False)

    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['A', 'B', 'C', 'D', 'E']
values = [1, 2, 3, 4, 5]
y_pos = np.arange(len(categories))

# Set up the figure and subplots
fig, axs = plt.subplots(1, 3, figsize=(14, 5))

# Plot 1: Complex background and 3D-style bars
axs[0].barh(y_pos, values, color='skyblue', edgecolor='blue', linewidth=2)
axs[0].set_yticks(y_pos)
axs[0].set_yticklabels(categories)
# axs[0].set_title('Overdesigned')
axs[0].imshow(np.linspace(0, 1, 256).reshape(1, -1),
              cmap='gray', aspect='auto',
              extent=[0, 5.5, -1, len(categories)], alpha=0.4)

# Plot 2: Simpler style with border
axs[1].barh(y_pos, values, color='lightgray', edgecolor='black', linewidth=1.5)
axs[1].set_yticks(y_pos)
axs[1].set_yticklabels(categories)
# axs[1].set_title('Improved Clarity')
# hide top and right spines
axs[1].spines['top'].set_visible(False)
axs[1].spines['right'].set_visible(False)

# Plot 3: Minimalist design
axs[2].barh(y_pos, values, color='lightgray')
axs[2].set_yticks(y_pos)
axs[2].set_yticklabels(categories)
# axs[2].set_title('Minimalist')
axs[2].grid(True, axis='x', linestyle='-', color='white', linewidth=2)
# hide spines
for spine in axs[2].spines.values():
    spine.set_visible(False)
axs[2].tick_params(axis='both', length=0)

# Common formatting
for ax in axs:
    ax.set_xlim(0, 5.5)
    ax.invert_yaxis()  # Optional: match top-down order

plt.tight_layout()
plt.show()
```
